#3. Buatlah visualisasi dengan mengunakan library python seperti matplotlib, seaborn atau tools lain

```python
#library yang digunakan
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# a. Countplot: Jumlah film/serial berdasarkan sertifikasi
plt.figure(figsize=(10, 6))
sns.countplot(x='Certificate', data=data)
plt.title('Jumlah Film/Serial Berdasarkan Sertifikasi')
plt.xlabel('Certificate')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# b. Pie Chart: Distribusi Genre (hanya 5 genre teratas)
genre_counts = data['Genre'].str.split(',').explode().value_counts().head(5)

plt.figure(figsize=(7, 7))
plt.pie(genre_counts, labels=genre_counts.index, autopct='%1.1f%%', startangle=90, colors=['#ffcc99', '#c2c2f0', '#ffb3e6', '#ff6666', '#99ff99'])
plt.title('Distribusi 5 Genre Teratas')
plt.axis('equal')
plt.show()
```
